```python
import torch

from torchvision.models.segmentation import deeplabv3_resnet101, DeepLabV3_ResNet101_Weights
```

```python
model = deeplabv3_resnet101(weights=DeepLabV3_ResNet101_Weights.DEFAULT).to("cuda")
```

```python
print(model)
```

```python
print(sum(p.numel() for p in model.parameters()))
```

```python
preprocess = DeepLabV3_ResNet101_Weights.DEFAULT.transforms()
print(preprocess)
batch = preprocess(torch.randn(2, 3, 300, 300)).to("cuda")
print(batch.shape)
# print(model(batch))
print(model(batch)['out'].detach().cpu().shape)
print(model(batch)['aux'].detach().cpu().shape)
```
