```python
import torch

from torchvision.models.segmentation import lraspp_mobilenet_v3_large, LRASPP_MobileNet_V3_Large_Weights
```

```python
model = lraspp_mobilenet_v3_large(weights=LRASPP_MobileNet_V3_Large_Weights.DEFAULT)
```

```python
print(model)
```

```python
print(sum(p.numel() for p in model.parameters()))
```

```python
preprocess = LRASPP_MobileNet_V3_Large_Weights.DEFAULT.transforms()
print(preprocess)
batch = preprocess(torch.randn(32, 3, 300, 300))
print(batch.shape)
print(model(batch)['out'].shape)
```
